Dataset information from READ.ME (Portuguese):

- A codificação de caracteres dos arquivos é "Latin 1";
- Os campos estão entre aspas e separados por ponto e vírgula, inclusive os campos numéricos;
- Campos preenchidos com #NULO significam que a informação está em branco no banco de dados. O correspondente para #NULO nos campos numéricos é -1;
- Campos preenchidos com #NE significam que naquele ano a informação não era registrada em banco de dados pelos sistemas eleitorais. O correspondente para #NE nos campos numéricos é -3;
- O campo UF, além das unidades da federação, pode conter alguma das seguintes situações:
o BR: quando se tratar de informação a nível nacional;
o VT: quando se tratar de voto em trânsito;
o ZZ: quando se tratar de Exterior.
- Os arquivos estão em constante processo de atualização e aperfeiçoamento. Alguns arquivos podem estar em branco ou com mensagem de erro devido a indisponibilidade temporária na base de algum estado ou à inexistência daquele arquivo para a época pretendida.

#### Now grabing just the information needed to understand the behaviour of Eduardo Paes votes

```python
df_paes = df[df["NM_VOTAVEL"] == "EDUARDO DA COSTA PAES"][[
    "NR_TURNO",
    "NR_ZONA",
    "NR_SECAO",
    "NM_VOTAVEL",
    "QT_VOTOS",
    "NR_LOCAL_VOTACAO",
    "NM_LOCAL_VOTACAO",
    "DS_LOCAL_VOTACAO_ENDERECO"
]]
```

```python
df_paes.tail(4)
```

```python
df_ze = pd.read_csv('data/csv-ze.csv', encoding='utf-8', delimiter=',')
df_ze.head(5)
```

```python
df_paes = df_paes.merge(
                        df_ze[['número da Zona','Bairro do local de votação']],
                        left_on='NR_ZONA', right_on='número da Zona',
                        how='left'
                        )

df_paes.head(2)
```

```python
df_paes.rename(columns={'Bairro do local de votação': 'NM_BAIRRO'}, inplace=True)
df_paes.drop(columns=['número da Zona'], inplace=True)
df_paes.head(2)
```

```python
df_paes["NM_BAIRRO"].unique()
```

### Creating a Dataframe (and a CSV) with the 2020 votes by Electoral zone

```python
#Drop
#Index no stack
#Group e soma

#Drop
df_temp = df_paes.drop(['NR_SECAO','NM_VOTAVEL', 'NR_LOCAL_VOTACAO', 'NM_LOCAL_VOTACAO', 'DS_LOCAL_VOTACAO_ENDERECO'], axis=1)
df_temp.head(2)
```

```python
#Index no stack
df_temp.set_index('NR_ZONA', inplace=True)
df_temp.head(2)
```

```python
#df_temp2 = df_temp.groupby(['NR_ZONA','NM_BAIRRO','NR_TURNO'])['QT_VOTOS'].sum()
df_temp2 = df_temp.groupby(['NR_ZONA','NR_TURNO'])['QT_VOTOS'].sum()
df_temp2.head(10)
```

```python
df_result_2020 = df_temp2.unstack(fill_value=0)
df_result_2020.head(10)
```
